```python
import pandas as pd

# Load the CSV file into a DataFrame
df = pd.read_csv('../Data/dsc_fc_summed_spectra_2023_v01.csv')

# List of time values you want to filter
time_list = ['00:00:00', '03:00:00', '06:00:00', '09:00:00', '12:00:00', '15:00:00', '18:00:00', '21:00:00']

# Convert the "Time" column to datetime
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M:%S')

# Filter rows where the "Time" column matches any value in the list
filtered_df = df[df['Time'].dt.strftime('%H:%M:%S').isin(time_list)]

# Save the filtered DataFrame to a new CSV file or perform further operations
filtered_df.to_csv('../Data/3hr-values.csv', index=False)
```
