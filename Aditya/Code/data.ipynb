```python
ref = {
    '00:00' : 'Kp1',
    '03:00' : 'Kp2',
    '06:00' : 'Kp3',
    '09:00' : 'Kp4',
    '12:00' : 'Kp5',
    '15:00' : 'Kp6',
    '18:00' : 'Kp7',
    '21:00' : 'Kp8',
}
```

```python
plasma = pd.read_csv('../Data/f_plasma.csv', engine='pyarrow')
kp = pd.read_csv('../Data/Kp-2023.csv', engine='pyarrow')
```

```python
plasma
```

```python
kp
```

```python
plasma['Hour'] = plasma['Time'].dt.strftime('%H:%M:%S')
plasma['Date'] = plasma['Time'].dt.strftime('%d-%m-%Y')
plasma
```

```python
plasma.info()
```

```python
# Iterate through Kp rows
for index, row in kp.iterrows():
    # Iterate through plasma rows
    for i, r in plasma.iterrows():
        # If the date and time match, add the Kp value to the plasma dataframe
        if row['Date'] == r['Date']:
            if r['Hour'] == '00:00:00':
                plasma.loc[i, 'Kp'] = row['Kp1']
            elif r['Hour'] == '03:00:00':
                plasma.loc[i, 'Kp'] = row['Kp2']
            elif r['Hour'] == '06:00:00':
                plasma.loc[i, 'Kp'] = row['Kp3']
            elif r['Hour'] == '09:00:00':
                plasma.loc[i, 'Kp'] = row['Kp4']
            elif r['Hour'] == '12:00:00':
                plasma.loc[i, 'Kp'] = row['Kp5']
            elif r['Hour'] == '15:00:00':
                plasma.loc[i, 'Kp'] = row['Kp6']
            elif r['Hour'] == '18:00:00':
                plasma.loc[i, 'Kp'] = row['Kp7']
            elif r['Hour'] == '21:00:00':
                plasma.loc[i, 'Kp'] = row['Kp8']
```

```python
plasma
```

```python
plasma.drop('Hour', axis=1, inplace=True)
plasma.drop('Date', axis=1, inplace=True)
```
